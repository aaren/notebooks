That's it. So the steps are:

- Take the logical or of the dx, dy gradients to find the
  edges of the masked region
  
- Order the points that make up the edge by the angle around
  their centre
  
- Create a `matplotlib.path.Path` instance with the ordered points

- Use the path instance to check whether any given point is valid. `&` this with the 2nd camera mask.
  This vectorises over arrays.
  

The valid points are necessarily flattened. How do we want to proceed with this? Is it sufficient to leave them all flattened (as we have coords and data anyway), or do we want to use a masked array?

We need to use a **masked array** as we'll have to regrid the data otherwise.

```python
def find_index_edges_of_domain(mask_image):
    """mask_image is a boolean array of validity over the
    measurement domain. This is True in the middle (where the
    data is valid) and False around the edges.
    
    This function finds the indices of the edges of the valid domain.
    """
    dx, dy = np.gradient(mask_image)
    is_edge = np.where(np.logical_or(dx, dy))
    return is_edge


def calculate_real_coords(mask_image, (scale_x, offset_x), (scale_y, offset_y)):
    """Convert the indices of mask_image to real coordinates,
    using a given scale and offset.
    """
    iy, ix = np.indices(mask_image.shape)
    sy, sx = mask_image.shape
    
    ry = scale_y * iy / sy - offset_y
    rx = scale_x * ix / sx - offset_x
    
    return ry, rx


def find_edges_of_domain(mask_image, convx, convy):
    """Calculate the edges of the valid domain in the mask_image
    using real coordinates, calculated using the (scale, offset)
    pairs given by convx and convy.
    
    Returns a (N, 2) array of the edge points.
    """
    ry, rx = calculate_real_coords(mask_image, convx, convy)
    is_edge = find_index_edges_of_domain(mask_image)
    edge_x = rx[is_edge]
    edge_y = ry[is_edge]
    return np.array((edge_x, edge_y)).T


mask_image_1 = imread('/home/eeaol/lab/data/flume2/main_data/image_limits/Define Mask - Reject(3nvpvi2w)/image_mask.3nvpvi2w.000000.bmp')[::-1,:].astype(np.bool)
mask_image_2 = imread('/home/eeaol/lab/data/flume2/main_data/image_limits/Define Mask - Reject(3nvpz1v5)/image_mask.3nvpz1v5.000000.bmp')[::-1,:].astype(np.bool)

scale_factor_1 = ((0.250052, 0.10244), (0.175165, 0.112465))
scale_factor_2 = ((0.235529, 0.11395), (0.167744, 0.114789))

edge_points_1 = find_edges_of_domain(mask_image_1, *scale_factor_1)
edge_points_2 = find_edges_of_domain(mask_image_2, *scale_factor_2)
def ccworder(A):
    A = A - mean(A, 1)[:, None]
    return argsort(arctan2(A[1, :], A[0, :]))

def test_is_valid(edge_points):
    """Create a method that tests whether given points lie
    within the edge defined by edge_points.
    
    Finds the convex hull of the points (guaranteed to be
    ordered in 2d), then makes these into 
    a matplotlib Path instance."""
    import scipy.spatial as spatial
    hull = spatial.ConvexHull(edge_points)
    ordered_edge = hull.points[hull.vertices]
    edge_path = mpl.path.Path(ordered_edge)
    return edge_path.contains_points

test_valid_1 = test_is_valid(edge_points_1)
test_valid_2 = test_is_valid(edge_points_2)
```

### Orientation

Ideally, the laser light sheet would be perfectly vertical and aligned with the central axis of the tank. In reality it is difficult to achieve this alignment and the light sheet is off axis.

There are two separate issues here:

1. calibration plane not equal to laser plane
2. laser plane off axis

The first is dealt with in Dynamic Studio. Data from an experiment is used to refine the initial calibration, with the assumption that they are from slightly different planes. The refined calibration is then used to produce the final data.

We can see that the laser is off axis by looking at the cross stream velocity component. As the experimental setup is symmetric about the cross stream, we expect this component of the velocity to have a symmetric distribution about a mean of zero.

If we just consider the valid region over the time that the current is coming through we get some good statistics.

```python
valid = np.logical_and(r.x[:] > 60, r.x[:] < 70) \
        & np.logical_and(r.z[:] > -100, r.z[:] < 20) \
        & np.logical_and(r.t[:] > 36, r.t[:] < 69)

nonzero = np.where((r.u[valid] != 0) & (r.v[valid] !=0) & (r.w[valid] != 0))
cross_stream = r.v[valid][nonzero]
down_stream = r.u[valid][nonzero]
vertical = r.w[valid][nonzero]
```

The cross stream velocity is clearly asymmetric.

A pure down stream velocity of magnitude $u$ will have a projection into a laser plane inclined at angle $\theta$ to the down stream axis of $(u \cos\theta, -v \sin\theta)$.

This angle $\theta$ is unknown to us and is what we need to determine to correct the system. We have the measured components $(u', v', w')$ and we wish to know the real components $(u, v, w)$. With the assumption that the light sheet is vertical ($w = w'$), we have the relation

$$
\begin{pmatrix}
u' \\ v' \\ w'
\end{pmatrix} = \begin{pmatrix}
\cos \theta & -\sin \theta & 0 \\
\sin \theta & \cos \theta & 0 \\
0 & 0 & 1 \end{pmatrix} \begin{pmatrix}
u \\ v \\ w 
\end{pmatrix}
$$

The inverse of a rotation matrix is equal to its transpose, i.e.

$$
\begin{pmatrix}
u \\ v \\ w
\end{pmatrix} = \begin{pmatrix}
\cos \theta & \sin \theta & 0 \\
-\sin \theta & \cos \theta & 0 \\
0 & 0 & 1 \end{pmatrix} \begin{pmatrix}
u' \\ v' \\ w'
\end{pmatrix}
$$

There are two ways of going about solving this problem. The simple thing to do is to just look at the means. The harder thing is to determine the transfer function for the probability distribution.

Just looking at the mean we seek $\theta$ such that $\bar v = 0$. We obtain

$$ \bar u = \bar u' \cos \theta + \bar v' \sin \theta $$
$$ \bar v = - \bar u' \sin \theta + \bar v' \cos \theta $$

We can use $ \bar v = 0 $, but it doesn't really get us anywhere as we don't know about the ideal pdf of $u$.

There are a couple of ways to approach the harder approach - exact and iterative. It is easier to iterate over a load of different angles until we converge on one that gives us a normal distribution for $w$. It is harder to create a theoretical transfer function for the ideal distribution of $w$, especially as we don't know what the ideal distributions of $u$ and $v$ should be.

Let's try the first approach - treating this as an optimisation problem.

For a given $\theta$ we compute $v = -u'\sin\theta + v'\cos\theta$ and fit a skew normal distribution. Vary $\theta$ to find zero skew.

```python
import scipy.stats as stats

#print stats.skewtest(cross_stream)
#print stats.skew(cross_stream)
#print stats.normaltest(cross_stream)

def real_cross_stream(theta):
    return -down_stream * np.sin(theta) + cross_stream * np.cos(theta)

def optimise_this(theta):
    return stats.normaltest(real_cross_stream(theta))

#print optimise_this(theta=-0.4)

plt.hist(real_cross_stream(theta=-0.01), bins=500, range=(-0.02, 0.02))
xlim(-0.02, 0.02)
```

Fiddling with the angle isn't solving the problem. It might be that we need to consider a two axis rotation.

The other axis that matters is the downstream axis and the rotation matrix about this axis is

$$
R_x = \begin{pmatrix}
1 & 0 & 0 \\
0 & \cos \phi & \sin \phi \\
0 & -\sin \phi & \cos \phi 
\end{pmatrix}
$$

Then for a given angle vector $(\phi, 0, \theta)$, we compute the matrix $ R = R_x.R_z $ and apply this to our data.

```python
def rotation_matrix(theta, phi=0):
    cx = np.cos(phi)
    sx = np.sin(phi)
    cz = np.cos(theta)
    sz = np.sin(theta)
    
    Rx = np.array([[1,   0,  0],
                   [0,  cx, sx],
                   [0, -sx, cx]])
    
    Rz = np.array([[ cz,  sz, 0],
                   [-sz,  cz, 0],
                   [  0,   0, 1]])
    
    return np.dot(Rx, Rz)
```

```python
def transform_cross_stream(theta, phi):
    rot = rotation_matrix(theta, phi)
    data = np.array((down_stream, cross_stream, vertical))
    rd, rc, rv = np.dot(rot, data)
    return rc

plt.hist(transform_cross_stream(-0.01, -0.01), bins=500, range=(-0.02, 0.02))
xlim(-0.02, 0.02)
```

### Non-dimensionalisation

We can't compare runs unless they are non dimensionalised. This rescales distances, speeds and times.

After rescaling, we have to re-grid the data so that all runs have a common grid. Use `scipy.interpolate.griddata` for this. Whether it is valid to do this depends on whether we have oversampled the velocity fluctuations in time and space.
