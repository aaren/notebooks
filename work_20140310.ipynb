Front fitting
-------------

Fit the front passage with a straight line. This is just one new
function `fit_front` on the `PreProcessor` which fits a straight
line to the front passage and resamples the time of passage from
that line:

```python
def detect_front(self):
    """Detect the trajectory of the gravity current front
    in space and time.

    Takes the column average of the vertical velocity and
    searches for the first time at which this exceeds a
    threshold.
    """
    threshold = 0.01  # m / s
    column_avg = self.W.mean(axis=0)
    exceed = column_avg > threshold
    # find first occurence of exceeding threshold
    front_it = np.argmax(exceed, axis=1)
    front_ix = np.indices(front_it.shape).squeeze()

    front_space = self.X[0][front_ix, front_it]
    front_time = self.T[0][front_ix, front_it]

    return front_space, front_time

def fit_front(self):
    """Fit a straight line to the detected front
    and resample the front times from this straight
    line.
    """
    front_space, front_time = self.detect_front()
    front_function = np.poly1d(np.polyfit(front_space, front_time, 1))
    straight_time = front_function(front_space)
    return front_space, straight_time
```

Alternate front relative transform
----------------------------------

We have been transforming the data to a front relative coordinate
system by using the front trajectory to define a skewed volume of
coordinates to extract from the data space and map to a regular
grid.

The volume of coordinates is defined relative to the *time* of front
passage, across all height and all horizontal positions.

We can equally well define the skewed volume of coordinates relative
to the *position* that the front passes at.

Let's create an alternate pre-processor with a new method that
computes the spatial transform:

```python
import scipy.ndimage as ndi
class AltPreProcessor(g.PreProcessor):
    def transform_to_front_relative_space(self, fit='1d'):
        if not fit:
            front_space, front_time = self.detect_front()
        elif fit == '1d':
            front_space, front_time = self.fit_front()

        rx = self.X[0, :, 0]
        dx = rx[1] - rx[0]

        x0 = -0.2
        x1 = 0.2
        relative_samples = np.arange(x0, x1, dx)
        sz, sx = self.T.shape[:2]
        # size of sampling axis
        ss = relative_samples.size
        # FIXME: below is wrong
        # might be working with the above new axes
        xrelative_samples = np.tile(relative_samples, (sz, sx, 1))

        rxf = front_space[None, ..., None] + xrelative_samples

        # grid coordinates of the sampling times
        # (has to be relative to what the time is at
        # the start of the data).
        x_coords = (rxf - rx[0]) / dx

        # z coords are the same as before
        z_coords = np.indices(x_coords.shape)[0]

        # t coords are front_time extended over the sampling volume
        # plus whatever else you want to be able to see
        rt = self.T[0, 0, :]
        xfront_time = np.tile(front_time[None, ..., None], (sz, 1, ss))
        t_coords = (xfront_time - rt[0]) / (rt[1] - rt[0])

        # required shape of the coordinates array is
        # (3, rz.size, rx.size, rt.size)
        coords = np.concatenate((z_coords[None], 
                                 x_coords[None], 
                                 t_coords[None]), axis=0)

        # there are now two x dimensions - the original and the
        # sampling
        self.Xss_ = xrelative_samples
        self.Xs_ = self.X[:, :, 0, None].repeat(ss, axis=-1)
        self.Zs_ = self.Z[:, :, 0, None].repeat(ss, axis=-1)
        self.Ts_ = xfront_time

        self.Us_ = ndi.map_coordinates(self.U, coords, cval=np.nan)
        self.Vs_ = ndi.map_coordinates(self.V, coords, cval=np.nan)
        self.Ws_ = ndi.map_coordinates(self.W, coords, cval=np.nan)
```

```python
app = AltPreProcessor(run=r)
app.extract_valid_region()
app.transform_to_lock_relative()
```

In the time based method we are limited in the width of each slice
by the total experiment time, which goes on for such a long time
after the front passage that it is not a limiting factor. The number
of slices is fixed at the experimental window width.

In the space based method we are limited in the width of each slice
by the experimental window *width*, which the front trajectory
completely traverses. There is a tradeoff between the width of each
slice - the window width about the front -  and the number of
slices.

In the example above we can imagine extracting more information
about the current by continuing the volume extraction beyond the
plot. This would require extrapolating from the front detection into
the region outside of the measurement window.

Statistical Homogeneity of a single run
---------------------------------------

Our conceptual model for the gravity current is a block of
statistics (features, turbulence) being advected by the bulk motion
of the current. We would like to see whether this is a valid idea
for what is going on.

Take the streamwise velocity and subtract the current velocity to
obtain the current relative motion. This should be zero mean in
statistically homogeneous regions.

We can get the speed of the front from the processed data like this:

```python
pr = g.ProcessedRun(g.default_processed + 'r13_12_17c.hdf5')

def compute_speed(fx, ft):
    dx = np.diff(fx)
    dt = np.diff(ft)
    return np.mean(dx / dt)
```
